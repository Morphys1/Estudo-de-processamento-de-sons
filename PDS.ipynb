```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Carregar o arquivo de áudio
file_path = '/717114__gregorquendel__we-wish-you-a-merry-christmas-arranged-for-strings.wav'  # Substitua pelo caminho do seu arquivo de áudio
audio_data, sr = librosa.load(file_path, sr=None)

# Definir o tempo de início e duração do trecho de interesse
start_time = 5  # Tempo inicial em segundos
duration = 1.0  # Duração do trecho em segundos

# Encontrar os índices dos samples correspondentes ao trecho de interesse
start_sample = int(start_time * sr)
end_sample = int((start_time + duration) * sr)
segment = audio_data[start_sample:end_sample]

# Calcular a frequência fundamental para o trecho selecionado
fft = np.fft.fft(segment)
magnitude = np.abs(fft)
frequency = np.fft.fftfreq(len(magnitude), d=1/sr)
fundamental_freq_index = np.argmax(magnitude)
fundamental_freq = np.abs(frequency[fundamental_freq_index])

# Gerar um sinal senoidal com a frequência fundamental encontrada
time = np.linspace(0, duration, len(segment))
sinusoid = np.sin(2 * np.pi * fundamental_freq * time)

# Determinar o número de períodos a serem exibidos (neste caso, 3 pulsos)
num_periods = 3

# Calcular o intervalo de tempo para exibir os 3 pulsos
time_interval = 1 / fundamental_freq * num_periods

# Plotar a forma de onda da frequência fundamental com o zoom nos 3 primeiros pulsos
plt.figure(figsize=(8, 4))
plt.plot(time, sinusoid)
plt.title('Forma de Onda dos Primeiros 3 Pulsos da Frequência Fundamental')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.xlim(0, time_interval)  # Definir o intervalo de tempo para mostrar os 3 pulsos
plt.show()
```

```python
# Os sinais são elementos de um espaço vetorial sobre o corpo dos números. Portanto vetores. Veja ilustrada abaixo a propriedade associativa.
#Pois um sinal correponde a uma função real do tempo(real) e pode se demonstrar que objetos desse tipo formam um espaço vetorial com escalar real.
# As operações são as adições de função e multiplicação por escalar usuais
import librosa
import numpy as np
import matplotlib.pyplot as plt

# Carregar o arquivo de áudio
file_path = '/content/717114__gregorquendel__we-wish-you-a-merry-christmas-arranged-for-strings.wav'  # Substitua pelo caminho do seu arquivo de áudio
audio_data, sr = librosa.load(file_path, sr=None)

# Definir o tempo de início e duração do trecho de interesse
start_time = 5  # Tempo inicial em segundos
duration = 1.0  # Duração do trecho em segundos

# Encontrar os índices dos samples correspondentes ao trecho de interesse
start_sample = int(start_time * sr)
end_sample = int((start_time + duration) * sr)
segment = audio_data[start_sample:end_sample]

# Calcular a FFT (Transformada Rápida de Fourier) para o trecho selecionado
fft = np.fft.fft(segment)
magnitude = np.abs(fft)
frequency = np.fft.fftfreq(len(magnitude), d=1/sr)

# Encontrar os índices dos picos no espectro (excluindo o DC)
non_dc_magnitude = magnitude[1:]  # Excluir o componente DC
non_dc_frequency = frequency[1:]  # Excluir o componente DC
top_indices = np.argsort(-non_dc_magnitude)[:100]  # Considerar os 100 maiores picos como candidatos

# Escolher aleatoriamente 3 picos distintos
chosen_indices = np.random.choice(top_indices, 3, replace=False)

# Encontrar as frequências dos 3 picos escolhidos
chosen_frequencies = np.abs(non_dc_frequency[chosen_indices])

# Gerar os sinais senoidais correspondentes às 3 frequências escolhidas
time = np.linspace(0, duration, len(segment))
signals = [np.sin(2 * np.pi * freq * time) for freq in chosen_frequencies]

# Ajuste para considerar apenas os três primeiros pulsos de cada sinal
pulse_length = int(sr / chosen_frequencies[0] * 3)  # Tamanho dos três primeiros pulsos

# Definir os três primeiros pulsos para cada sinal
signal_1_pulse = signals[0][:pulse_length]
signal_2_pulse = signals[1][:pulse_length]
signal_3_pulse = signals[2][:pulse_length]

# Calcular as expressões conforme solicitado
expression_1 = (signal_1_pulse + signal_2_pulse) + signal_3_pulse
expression_2 = signal_1_pulse + (signal_2_pulse + signal_3_pulse)
expression_3 = signal_1_pulse + signal_2_pulse
expression_4 = signal_2_pulse + signal_1_pulse

# Plotar as expressões separadamente
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(time[:pulse_length], expression_1)
plt.title('(Sinal 1 + Sinal 2) + Sinal 3 - Três primeiros pulsos')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(time[:pulse_length], expression_2)
plt.title('Sinal 1 + (Sinal 2 + Sinal 3) - Três primeiros pulsos')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
#Ilustrada a distribtvidade da multiplicação por escalar em relação a soma de sinais(vetores)

# Carregar o arquivo de áudio
file_path = '/content/717114__gregorquendel__we-wish-you-a-merry-christmas-arranged-for-strings.wav'  # Substitua pelo caminho do seu arquivo de áudio
audio_data, sr = librosa.load(file_path, sr=None)

# Definir o tempo de início e duração do trecho de interesse
start_time = 5  # Tempo inicial em segundos
duration = 1.0  # Duração do trecho em segundos

# Encontrar os índices dos samples correspondentes ao trecho de interesse
start_sample = int(start_time * sr)
end_sample = int((start_time + duration) * sr)
segment = audio_data[start_sample:end_sample]

# Calcular a FFT (Transformada Rápida de Fourier) para o trecho selecionado
fft = np.fft.fft(segment)
magnitude = np.abs(fft)
frequency = np.fft.fftfreq(len(magnitude), d=1/sr)

# Encontrar os índices dos picos no espectro (excluindo o DC)
non_dc_magnitude = magnitude[1:]  # Excluir o componente DC
non_dc_frequency = frequency[1:]  # Excluir o componente DC
top_indices = np.argsort(-non_dc_magnitude)[:100]  # Considerar os 100 maiores picos como candidatos

# Escolher aleatoriamente 2 picos distintos
chosen_indices = np.random.choice(top_indices, 2, replace=False)

# Encontrar as frequências dos 2 picos escolhidos
chosen_frequencies = np.abs(non_dc_frequency[chosen_indices])

# Gerar os sinais senoidais correspondentes às 2 frequências escolhidas
time = np.linspace(0, duration, len(segment))
signals = [np.sin(2 * np.pi * freq * time) for freq in chosen_frequencies]

# Ajuste para considerar apenas os três primeiros pulsos de cada sinal
pulse_length = int(sr / chosen_frequencies[0] * 3)  # Tamanho dos três primeiros pulsos

# Multiplicar cada frequência por um escalar
scalar = 2  # Mesmo escalar para ambos os processos
scaled_signals = [scalar * signal[:pulse_length] for signal in signals]

# Somar os valores dos sinais escalados
sum_scaled_signals = sum(scaled_signals)

# Realizar o produto de um escalar com a soma dos sinais originais
product_sum_signals = scalar * sum(signals)[:pulse_length]

# Plotar os três primeiros pulsos dos resultados finais
plt.figure(figsize=(8, 4))

plt.subplot(2, 1, 1)
plt.plot(time[:pulse_length], sum_scaled_signals)
plt.title('Três primeiros pulsos do resultado da soma dos produto dos sinais originais com o escalar')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(time[:pulse_length], product_sum_signals)
plt.title('Três primeiros pulsos do resultado do produto do escalar com a soma dos sinais originais')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
